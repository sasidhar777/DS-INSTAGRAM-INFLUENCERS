```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv(r"Influencer.csv")
```

```python
df
```

```python
Follower=[]
for element in df['Followers']:
    str1=0
    #print(element)
    if element[-1]=='k':
        #print(float(element[0:len(element)-1]))
        str1=float(element[0:len(element)-1])
        Follower.append(str1*1000)
        #print(str1)
    elif element[-1]=='m':
        str1=float(element[0:len(element)-1])
        Follower.append(str1*1000000)
        #print(str1)
    elif element[-1]=='b':
        str1=float(element[0:len(element)-1])
        Follower.append(str1*1000000000)
    else:
        str1=float(element)
        Follower.append(str1)
print(Follower)  
```

```python
Likes=[]
for element in df['Avg. Likes']:
    str1=0
    #print(element)
    if element[-1]=='k':
        #print(float(element[0:len(element)-1]))
        str1=float(element[0:len(element)-1])
        Likes.append(str1*1000)
        #print(str1)
    elif element[-1]=='m':
        str1=float(element[0:len(element)-1])
        Likes.append(str1*1000000)
        #print(str1)
    elif element[-1]=='b':
        str1=float(element[0:len(element)-1])
        Likes.append(str1*1000000000)
    else:
        str1=float(element)
        Likes.append(str1)
print(Likes)  
```

```python
Post=[]
for element in df['Posts']:
    str1=0
    #print(element)
    if element[-1]=='k':
        #print(float(element[0:len(element)-1]))
        str1=float(element[0:len(element)-1])
        Post.append(str1*1000)
        #print(str1)
    elif element[-1]=='m':
        str1=float(element[0:len(element)-1])
        Post.append(str1*1000000)
        #print(str1)
    elif element[-1]=='b':
        str1=float(element[0:len(element)-1])
        Post.append(str1*1000000000)
    else:
        str1=float(element)
        Post.append(str1)
print(Post)  
```

```python
New_post_likes=[]
for element in df['New Post Avg. Likes']:
    str1=0
    #print(element)
    if element[-1]=='k':
        #print(float(element[0:len(element)-1]))
        str1=float(element[0:len(element)-1])
        New_post_likes.append(str1*1000)
        #print(str1)
    elif element[-1]=='m':
        str1=float(element[0:len(element)-1])
        New_post_likes.append(str1*1000000)
        #print(str1)
    elif element[-1]=='b':
        str1=float(element[0:len(element)-1])
        New_post_likes.append(str1*1000000000)
    else:
        str1=float(element)
        New_post_likes.append(str1)
print(New_post_likes)  
```

```python
Total_likes=[]
for element in df['Total Likes']:
    str1=0
    #print(element)
    if element[-1]=='k':
        #print(float(element[0:len(element)-1]))
        str1=float(element[0:len(element)-1])
        Total_likes.append(str1*1000)
        #print(str1)
    elif element[-1]=='m':
        str1=float(element[0:len(element)-1])
        Total_likes.append(str1*1000000)
        #print(str1)
    elif element[-1]=='b':
        str1=float(element[0:len(element)-1])
        Total_likes.append(str1*1000000000)
    else:
        str1=float(element)
        Total_likes.append(str1)
print(Total_likes) 
```

```python
Channel_Info=[]
for original_string in df['Channel Info']:
        replacement_string = original_string.replace("\n", "")
        Channel_Info.append(replacement_string)

Channel_Info
```

```python
correlation_matrix = df2.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
correlation_matrix
```

```python
# Find the highest correlation coefficient
max_corr_1 = correlation_matrix.values.max()

# Set the maximum value to a small value (e.g., negative infinity)
correlation_matrix.values[correlation_matrix.values == max_corr_1] = -np.inf

# Find the second-highest correlation coefficient
max_corr = correlation_matrix.values.max()

# Find the indices of the highest correlation in the matrix
row, col = np.where(correlation_matrix.values == max_corr)

# Get the names of the columns with the highest correlation
column1_name = df.columns[row[0]]
column2_name = df.columns[col[0]]

print(f"Highest correlation is between '{column1_name}' and '{column2_name}' with a correlation coefficient of {max_corr:.2f}")
```

```python
import matplotlib.pyplot as plt

# Set up the subplots for the histograms
plt.figure(figsize=(12, 8))

# Frequency Distribution for Influence Score
plt.subplot(3, 1, 1)
plt.hist(df1['Influence Score'], bins=9, color='blue', edgecolor='black')
plt.xlabel('Influence Score')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Influence Score')

# Frequency Distribution for Followers
plt.subplot(3, 1, 2)
plt.hist(df1['Followers'], bins=5, color='green', edgecolor='black')
plt.xlabel('Followers')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Followers')

# Frequency Distribution for Posts
plt.subplot(3, 1, 3)
plt.hist(df1['Posts'], bins=7, color='red', edgecolor='black')
plt.xlabel('Posts')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Posts')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set up the subplots for the histograms
plt.figure(figsize=(15, 8))

# Function to plot histograms with value annotations
def plot_histogram_with_annotations(data, x_label, y_label, title, num_bins, color):
    counts, bins, patches = plt.hist(data, bins=num_bins, color=color, edgecolor='black')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    for count, x, patch in zip(counts, bins, patches):
        plt.text(x + (bins[1] - bins[0]) / 2, count, str(int(count)), ha='center', va='bottom')

# Frequency Distribution for Influence Score
plt.subplot(3, 1, 1)
plot_histogram_with_annotations(df1['Influence Score'], 'Influence Score', 'Frequency', 'Frequency Distribution of Influence Score', 9, 'blue')

# Frequency Distribution for Followers
plt.subplot(3, 1, 2)
plot_histogram_with_annotations(df1['Followers'], 'Followers', 'Frequency', 'Frequency Distribution of Followers', 5, 'green')

# Frequency Distribution for Posts
plt.subplot(3, 1, 3)
plot_histogram_with_annotations(df1['Posts'], 'Posts', 'Frequency', 'Frequency Distribution of Posts', 7, 'red')

plt.tight_layout()
plt.show()
```
